```python
clarity_dict = {"IF": 7, "VVS1":6, "VVS2":5,"VS1":4,"VS2":3,"SI2":1,"I1":0}
data['clarity'].replace(clarity_dict, inplace=True)
data.head()
```

```python
data = pd.get_dummies(data, columns=['cut', 'color'])
data.head()
```

```python
NUMERIC_FEATURES = ['carat', 'depth', 'table', 'x', 'y', 'z']
numeric_df = data[NUMERIC_FEATURES]
numeric_df.describe()
```

```python
#standardize numeric data by subtracting the mean and dividing by the std dev. This is done by scale of sklearn
from sklearn.preprocessing import scale
numeric_array = scale(numeric_df)
numeric_array[:5]
```

```python
numeric_df = pd.DataFrame(numeric_array, columns=numeric_df.columns)
numeric_df.describe()
```

```python
data.head()
```

```python
data.drop(NUMERIC_FEATURES, inplace=True, axis=1)
data.head()
```

```python
data = pd.concat([data, numeric_df], axis=1)
data.shape
```

```python
data.head()
```

```python
data.to_csv('datasets/diamonds_processed_self.csv')
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('price', axis=1)
Y = data['price']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
```
