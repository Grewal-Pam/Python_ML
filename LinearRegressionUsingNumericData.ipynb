```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('medv', axis = 1)
Y = data['medv']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state=0)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
linearRegression = LinearRegression()
linearRegression.fit(x_train    , y_train)
```

```python
x_test.age
```

```python
y_pred = linearRegression.predict(x_test)
y_pred
```

```python
#lets understand the details behind the prediction
#Linear regression = > y= mx + b
# m is slope or gradient
# b is the y intercept(point that crossed the y axis)
# x and y are the values eg- age and medv for this example.
#x independent variables or features
#y is dependent variable
```

```python
#coef_ gives the value of m slope or gradient
linearRegression.coef_
```

```python
#intercept gives value of b
linearRegression.intercept_
```

```python
#y=m*x+b 
-4.62321394e-03*46.7+44.966308953062345
```

```python
df_y = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
df_y.sample(10)
```

```python
print("Training Score(r square): ", linearRegression.score(x_train, y_train))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print("Testing score: ", score)
```
